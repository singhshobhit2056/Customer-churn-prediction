```python
#change datatype of total charges to int64
#drop serial number customer id
df = df.drop("customerID", axis = 'columns')
```

```python
df[pd.to_numeric(df.TotalCharges,errors = 'coerce').isnull()] # it says leave the errors as it is
```

```python
df1 = df[df.TotalCharges != ' ']
```

```python
df1.shape
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
df1
```

```python
df1.OnlineBackup.unique()
```

```python
df1.dtypes
```

```python
df1['gender'].replace(['Male', 'Female'], [1,0], inplace = True)
df1['Partner'].replace(['Yes', 'No'], [1,0], inplace = True)
df1['Dependents'].replace(['Yes', 'No'], [1,0], inplace = True)
df1['PhoneService'].replace(['Yes', 'No'], [1,0], inplace = True)
df1['MultipleLines'].replace(['Yes', 'No','No phone service'], [1,0,0], inplace = True)
df1['InternetService'].replace(['DSL', 'Fiber optic','No'], [1,2,3], inplace = True)
df1['OnlineSecurity'].replace(['Yes', 'No', 'No internet service'], [1,0,0], inplace = True)
df1['DeviceProtection'].replace(['No', 'Yes', 'No internet service'], [0,1,0], inplace = True)
df1['TechSupport'].replace(['No', 'Yes', 'No internet service'], [0,1,0], inplace = True)
df1['StreamingTV'].replace(['No', 'Yes', 'No internet service'], [0,1,0], inplace = True)
df1['StreamingMovies'].replace(['Yes', 'No', 'No internet service'], [1,0,0], inplace = True)
df1['Contract'].replace(['Month-to-month', 'One year', 'Two year'], [1,2,3], inplace = True)
df1['PaperlessBilling'].replace(['Yes', 'No'], [1,0], inplace = True)
df1['PaymentMethod'].replace(['Electronic check', 'Mailed check', 'Bank transfer (automatic)','Credit card (automatic)'], [1,2,3,4], inplace = True)
df1['OnlineBackup'].replace(['Yes', 'No', 'No internet service'], [1,0,0], inplace = True)
df1['Churn'].replace(['No', 'Yes'], [0,1], inplace = True)

```

```python
df1
```

```python
df1.dtypes
```

```python
x = df1.drop('Churn', axis = 'columns')
y = df1['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
y_train
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(20, input_shape = (19,), activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid'),
    ])

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

model.fit(x_train, y_train, epochs = 100)
```

```python
model.evaluate(x_test,y_test)
```

```python
y1 = model.predict(x_test)

y_pred = []

for i in y1:
  if i>=0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(y_test, y_pred))
```
